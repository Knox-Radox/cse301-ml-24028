```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./dataset/Salary_Data.csv')
data.head()
```

```python
data.isnull().sum()
data = data.dropna()
```

```python
x = data.iloc[:, 4:5].values
y = data.iloc[:, 5].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
# print(y_pred)
# print(y_test)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Training Set
plt.scatter(x_train, y_train, color='blue', label='Actual Salary')
plt.plot(x_train, model.predict(x_train), color='red', label='Regression Line')
plt.title('Salary vs Years of Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
#Test Set
plt.scatter(x_test, y_test, color='blue', label='Actual Salary')
plt.plot(x_test, model.predict(x_test), color='red', label='Regression Line')
plt.title('Salary Prediction (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(x_test)

plt.scatter(y_test, y_pred)
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary")
plt.show()

# If points are close to the diagonal, the model is performing well.
```

```python
y_pred = model.predict(x_test)
residuals = y_test - y_pred

plt.scatter(y_pred, residuals)
plt.axhline(y=0)
plt.xlabel("Predicted Salary")
plt.ylabel("Residual")
plt.title("Residual Plot")
plt.show()
```
